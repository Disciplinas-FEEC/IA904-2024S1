```python
import os
import shutil

# Defina o caminho das pastas de origem e destino
pasta_origem = 'gdrive/MyDrive/Grupo IA904/Dia 31/'
pasta_destino = 'gdrive/MyDrive/Grupo IA904/Dia 031/'

# Função para renomear os arquivos
def renomear_arquivos(pasta_origem, pasta_destino):
    # Inicializa o contador para o nome dos arquivos
    contador = 0

    # Itera sobre cada pasta dentro da pasta de origem
    for pasta in os.listdir(pasta_origem):
        # Verifica se é uma pasta
        if os.path.isdir(os.path.join(pasta_origem, pasta)):
            print(f"Processando a pasta: {pasta}")
            # Itera sobre cada arquivo dentro da pasta
            for arquivo in os.listdir(os.path.join(pasta_origem, pasta)):
                # Cria o novo nome do arquivo
                novo_nome = f'frame_{contador:04}.jpg'
                # Move o arquivo para a pasta de destino com o novo nome
                shutil.move(os.path.join(pasta_origem, pasta, arquivo), os.path.join(pasta_destino, novo_nome))
                # Incrementa o contador
                contador += 1
    print("Processo concluído.")

# Chama a função para renomear e mover os arquivos
renomear_arquivos(pasta_origem, pasta_destino)
```
