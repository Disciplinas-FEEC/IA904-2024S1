```python
import pandas as pd
from mtcnn import MTCNN
import cv2
```

```python
df_east_asian=df[df['race']=='East Asian']
df_indian=df[df['race']=='Indian']
df_black=df[df['race']=='Black']
df_white=df[df['race']=='White']
df_middle_eastern=df[df['race']=='Middle Eastern']
df_latino_hispanic=df[df['race']=='Latino_Hispanic']
df_southeast_asian=df[df['race']=='Southeast Asian']
```

```python
img = cv2.cvtColor(cv2.imread("../data/processed/train/1.jpg"), cv2.COLOR_BGR2RGB)
detector = MTCNN()
detector.detect_faces(img)
```

```python
def get_confidences(image_path):
    img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    detector = MTCNN()
    faces = detector.detect_faces(img)
    confidences = [face['confidence'] for face in faces]
    return confidences
```

```python
df_race=df_southeast_asian
```

```python
df_race['confidence']=pd.NA
```

```python
df_race
```

```python
for i, image_path_short in enumerate(df_race['file']):
    image_path_long='../data/processed/'+image_path_short
    try:
        list_conf=get_confidences(image_path_long)
        if len(list_conf)==1:
            df_race.loc[i, 'confidence'] = float(list_conf)
        elif len(list_conf)==0:
            df_race.loc[i, 'confidence'] = 'nenhum rosto detectado (=0)'
        else:
            df_race.loc[i, 'confidence'] = 'mais de um rosto detectado'
    except:
        df_race.loc[i, 'confidence'] = 'não foi detectado nenhum rosto'
    if i==2000: 
        break
```

```python
df_race.to_csv('df_southeast_asian_to_2000.csv')
```
