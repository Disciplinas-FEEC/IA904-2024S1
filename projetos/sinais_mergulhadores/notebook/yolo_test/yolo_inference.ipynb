```python
import torch
from ultralytics import YOLO
import os
import cv2
from PIL import Image

# Verificar se a GPU está disponível
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"Usando GPU: {torch.cuda.get_device_name(0)}")
else:
    device = torch.device("cpu")
    print("Usando CPU")
```

```python
# Carregar o modelo YOLOv8 treinado
model_path = r'C:\Users\rapha\OneDrive\Documentos\mestrado\projeto_de_visao_computacional\Project\runs\detect\train\weights\best.pt'
print(f"Carregando o modelo treinado de {model_path}...")
model = YOLO(model_path)
```

```python
# Caminho para o conjunto de dados de teste
test_images_folder = r'c:\Users\rapha\OneDrive\Documentos\mestrado\projeto_de_visao_computacional\Project\Dataset\images\test'
output_folder = 'inference_results'

# Criar a pasta de saída, se não existir
os.makedirs(output_folder, exist_ok=True)
```

```python
# Realizar Inferência e Salvar Resultados
for img_name in os.listdir(test_images_folder):
    img_path = os.path.join(test_images_folder, img_name)
    img = Image.open(img_path)

    # Realizar inferência
    results = model(img, device=device)

    # Processar e salvar cada resultado
    for result in results:
        # Plota as detecções na imagem
        result_img = result.plot()

        # Converte a imagem plotada para o formato PIL
        result_pil = Image.fromarray(result_img)

        # Salva a imagem resultante na pasta de saída
        result_pil.save(os.path.join(output_folder, img_name))

    print(f"Resultados salvos para {img_name}")
```
