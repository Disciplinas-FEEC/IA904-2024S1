```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data_path = os.path.join(os.path.dirname(os.getcwd()), 'data', 'raw', 'OAK-D Synthetic Pose Dataset')

dataframe_info = pd.read_csv(os.path.join(data_path, 'DatasetInfo.csv'))
```

```python
def histogram(label):
    plt.hist(dataframe_info[label], edgecolor='black')
    plt.xlabel(label.capitalize())
    plt.ylabel('Frequence')
    plt.title('{} Histogram'.format(label.capitalize()))
    plt.show()
```

```python
histogram('age')
histogram('sex')
histogram('ethnicity')
histogram('height')
histogram('weight')
```
